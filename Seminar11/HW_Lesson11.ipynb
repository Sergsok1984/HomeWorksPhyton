```python
# Даны две функции:
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# 1. Найти координаты точек пересечения
# 2. Построить графики функций в одной системе координат
# 3. Построить график функции пересечения

from sympy import *

x = Symbol('x')
f = x**3 - 50*x
g = -1*x**4 + 88*x**2 - 241
h = 1*x**4 + x**3 - 88*x**2 - 50*x  + 241
```

```python
# 1. Координаты точек пересечения

dot_cross_x = solve(h, x)
dot_cross_y = list()
for i in range(len(dot_cross_x)):
    dot_cross_y.append(f.subs(x, dot_cross_x[i]))
    print(f'x{i+1}: {dot_cross_x[i]} \ny{i+1}: {dot_cross_y[i]}\n')
```

```python
#2. Графики функций в одной системе координат

plot(f, g)
```

```python
# 3. График функции пересечения

plot(h)
```
